```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pdb
from matplotlib.collections import (
    Collection, CircleCollection, LineCollection, PathCollection,
    PolyCollection, RegularPolyCollection)
from matplotlib.container import ErrorbarContainer, BarContainer, StemContainer
from matplotlib.patches import (Patch, Rectangle, Shadow, FancyBboxPatch,
                                StepPatch)
from matplotlib.lines import Line2D
```

```python
fig1, ax1 = plt.subplots(1,2,figsize = (10,5))

ax1[0].plot([0,1],[0,1],
        color = 'k',
        linewidth = 2)

ax1[0].plot([1,0],[0,1],
        color = 'b',
        linewidth = 2)

ax1[0].hist(np.random.uniform(0,1,15),
           bins = 10, density = True)

ax1[1].scatter([0,1,2], [0,1,2],
              color = 'purple',
              marker = 'o')

ax1[1].bar(x = [0,1,2],
          height = [0.25, 0.5, 0.75],
          width = 0.8)

ax1[1].errorbar([2,3,4], [0,1,2],
                 yerr = [0.1,0.1,0.1],
                 fmt = 'o',
                 capsize = 2,
                 label = 'silly')



ax1[0].set_xlabel('X', fontsize = 14)
ax1[0].set_ylabel('Y', fontsize = 14)
ax1[1].set_xlabel('X', fontsize = 14)
ax1[1].set_ylabel('Y', fontsize = 14)
ax1[0].grid()
ax1[1].grid()

```

```python
# Create the ExtractedFigure class

extractedfig = efc.ExtractedFigure(fig = fig1)
```

The artists are now in a list of lists.

The number of lists in ExtractedFigure.extracted_artist_list is the number of subplots.

The length of each subplot list is the number of artists plotted in each plot.

So, if a histogram was plotted in the first subplot as the third item plotted, then it would be called as:
extractedfig.extracted_artist_list[0][2]

```python
# Change the color of the histogram in the first subplot to orange
extractedfig.extracted_artist_list[0][2].set_color('orange')

extractedfig.extracted_figure
```
